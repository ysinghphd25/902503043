```python
import random
import csv
```

Generate Random Data File

```python
rows = 200
cols = 100

with open("datafile.csv", "w", newline="") as file:
    writer = csv.writer(file)

    for i in range(rows):
        row = [random.randint(1, 1000) for j in range(cols)]
        writer.writerow(row)

print("Data file created successfully")
```

Q2 — Assign Subsequent Rows to First Row Elements

Step 1 — Create Matrix

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
```

Step 2 — Assign Rows

```python
first_row = matrix[0]

for row in matrix[1:]:
    for i in range(len(first_row)):
        first_row[i] += row[i]

print(first_row)
```

Q3 — Group Similar Elements into Matrix

Step 1 — Input List

```python
lst = [1,1,2,2,3,3,4,4]
```

Step 2 — Group Elements

```python
matrix = []

for i in range(0, len(lst), 2):
    matrix.append([lst[i], lst[i+1]])

print(matrix)
```

Q4 — Add and Subtract Matrices

```python
A = [
    [1,2],
    [3,4]
]

B = [
    [5,6],
    [7,8]
]
```

Step 2 — Addition

```python
addition = []

for i in range(len(A)):
    row = []

    for j in range(len(A[0])):
        row.append(A[i][j] + B[i][j])

    addition.append(row)

print("Addition:")
print(addition)
```

Step 3 — Subtraction

```python
subtraction = []

for i in range(len(A)):
    row = []

    for j in range(len(A[0])):
        row.append(A[i][j] - B[i][j])

    subtraction.append(row)

print("Subtraction:")
print(subtraction)
```

Q5 — Create n × n Matrix with Even Opposite Corner Sum

Step 1 — Input Size

```python
n = 4
```

Step 2 — Generate Matrix

```python
matrix = []

num = 2

for i in range(n):
    row = []

    for j in range(n):
        row.append(num)
        num += 2

    matrix.append(row)

for row in matrix:
    print(row)
```
