```python
import numpy as np
import pandas as pd
```

# Part I: Randomized Sales Data Generation

This section generates random monthly sales data using NumPy and converts it into a Pandas DataFrame.

## Step 1: Initialize Random Seed

The roll number is used as a random seed so that each student gets unique sales data.

```python
np.random.seed(902503043)
```

## Step 2: Generate Sales Data

Random sales data is generated for 12 months and 4 product categories:
- Electronics
- Clothing
- Home & Kitchen
- Sports

Sales values range between 1000 and 5000 units.

```python
sales_data = np.random.randint(1000, 5001, size=(12,4))

print(sales_data)
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

## Step 3: Convert NumPy Array into Pandas DataFrame

The generated sales data is converted into a Pandas DataFrame using month names as index labels and product categories as column names.

```python
df = pd.DataFrame(
    sales_data,
    columns=['Electronics', 'Clothing', 'Home & Kitchen', 'Sports'],
    index=months
)

print(df)
```
