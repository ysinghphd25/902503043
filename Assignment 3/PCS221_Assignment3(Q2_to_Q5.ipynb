```python
import numpy as np
```

Create Array

```python
array = np.array([[1, -2, 3],
                  [-4, 5, -6]])

print("Original Array:")
print(array)
```

Q2(i) — Element-wise Absolute Value

```python
print("Element-wise Absolute Value:")

print(np.abs(array))
```

Q2(ii) — Percentiles

You need:

25th percentile
50th percentile
75th percentile

for:

Flattened array
Each column
Each row

```python
print("Flattened Array Percentiles:")

print(np.percentile(array, [25, 50, 75]))
```

 Column-wise Percentiles

```python
print("Column-wise Percentiles:")

print(np.percentile(array, [25, 50, 75], axis=0))
```

Row-wise Percentiles

```python
print("Row-wise Percentiles:")

print(np.percentile(array, [25, 50, 75], axis=1))
```

Q2(iii) — Mean, Median and Standard Deviation

```python
print("Flattened Array Statistics:")

print("Mean =", np.mean(array))
print("Median =", np.median(array))
print("Standard Deviation =", np.std(array))
```

Column-wise Statistics

```python
print("Column-wise Statistics:")

print("Mean:")
print(np.mean(array, axis=0))

print("Median:")
print(np.median(array, axis=0))

print("Standard Deviation:")
print(np.std(array, axis=0))
```

Row-wise Statistics

```python
print("Row-wise Statistics:")

print("Mean:")
print(np.mean(array, axis=1))

print("Median:")
print(np.median(array, axis=1))

print("Standard Deviation:")
print(np.std(array, axis=1))
```

# Q5: Swap Two Elements in a Set by Converting it into a List

```python
my_set = {1, 2, 3, 4}

print("Original Set:")
print(my_set)
```

```python
temp_list = list(my_set)

print("Converted List:")
print(temp_list)
```

```python
temp = temp_list[0]

temp_list[0] = temp_list[1]

temp_list[1] = temp
```

```python
my_set = set(temp_list)

print("Set After Swapping:")
print(my_set)
```
